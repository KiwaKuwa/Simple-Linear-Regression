```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

#โหลด data set สำหรับใช้ในการศึกษา
iris = sns.load_dataset('iris')

#ทำให้เหลือแค่ข้อมูลในส่วนที่เราต้องการ
iris = iris[['petal_length', 'petal_width']]
print(iris)

#กำหนดค่าตัวแปล
x = iris['petal_length']
y = iris['petal_width']

#ทำให้ x และ y เป็น array แบบ 2 มิติ
x = x.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)

#ลองพล็อตกราฟ
plt.scatter(x, y)
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
#แบ่งข้อมูลสำหรับ train และ test
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=0)

#แปลงให้เป็น array แบบ 2 มิติ
xtrain = xtrain.reshape(-1, 1)
ytrain = ytrain.reshape(-1, 1)
xtest = xtest.reshape(-1, 1)
ytest = ytest.reshape(-1, 1)

model = LinearRegression()
model.fit(xtrain, ytrain)

m = model.coef_
c = model.intercept_

#ลองให้ model ทำนายค่า y
ypredtrain = model.predict(xtrain)
plt.scatter(xtrain, ytrain)
plt.plot(xtrain, ypredtrain, color='red')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```
